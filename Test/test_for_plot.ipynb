```python
class ShapedPlot:
    styles = plt.style.available

    def __init__(self, notes: dict = None, figsize: tuple = None, subplots: tuple = None, style_num: int = 19):
        self.notes = notes
        if figsize is None:
            figsize = (6, 3)
        self.figsize = figsize
        if subplots is None:
            subplots = (1,1)
        self.subplots = subplots
        self._init_plot()
        plt.style.use(self.styles[style_num])
        

    def _init_plot(self):
        self.fig, self.axes = plt.subplots(*self.subplots, figsize=self.figsize)


    def clear(self):
        self.fig.clf()
```

```python
my_plot = ShapedPlot()
```

```python
my_plot.fig
```

```python
my_plot.clear()
```

```python
my_plot.axes.set_xlim(None)

```
