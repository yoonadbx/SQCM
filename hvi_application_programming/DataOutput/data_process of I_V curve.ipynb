```python
with open('measured_data_1.npy', 'rb') as f:
    data = np.load(f)
```

```python
plt.plot(data)
```

```python
plt.plot(data[20000:60000])
```

```python
data_smooth = sig.savgol_filter(data, window_length=5001, polyorder=3)
```

```python
plt.plot(data_smooth)
```

```python
plt.plot(data_smooth[20000:80000]/10000/1.5)
```

```python
chop = 20000
voltage_list = []
for ii in range(22):
    voltage_list.append(np.mean(data_smooth[chop*ii:chop*(ii+1)]))
    print(voltage_list[ii])
```

```python
repetitive_unit = np.linspace(1, 8, 20).reshape((1, 20))
start_unit = np.array([[0.5]])
end_unit = np.array([[0]])
voltage_input = np.hstack((start_unit, repetitive_unit, end_unit)).reshape(22)
```

```python
plt.scatter(voltage_input, voltage_list)
```
